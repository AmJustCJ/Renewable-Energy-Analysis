### Q1: Which Country Generate the most energy from 2000 to 2020, what type of renewable energy is it?


First, we concatenate all 4 top five countries that produce most energy from 2000 to 2020<br>
To prevent duplicate data, drop_duplicates() were used

````python
top_energy_df = pd.concat([most_avg_wind_df,most_avg_hydro_df,most_avg_solar_df,most_avg_other_df]).drop_duplicates().reset_index(drop=True)
top_energy_df
````

**Generate** a horizontal barplot, with x axis being countries and y axis being the energy produce, with each color bar represent each type of renewable energy

````python
top_energy_df.plot(x='Country', y=['Electricity from wind (TWh)',
                                   'Electricity from hydro (TWh)',
                                   'Electricity from solar (TWh)',
                                   'Electricity from other renewables including bioenergy (TWh)'],
                   kind = 'barh', figsize=(15,15), color=["limegreen", "royalblue", "orangered", "darkviolet"],
                   title = 'Top average energy production countries\n from 2000 to 2020')
plt.xlabel('Energy produced (TWh)')
plt.ylabel('Country')
plt.show();
````

<h3>From the horizontal bar chart above, we can conclude that China produce the most renewable energy on average from 2000 to 2020, and Hydro energy was the highest out of all other type of renewable energy

### Q2: What is the distribution of renewable energy production of the world in year 2021 (use pie chart)

First, create a dataframe that only take 'World' from "Entity" column that is from year 2021

````python
world_prod_df = renewable_energy_prod_df[(renewable_energy_prod_df['Entity'] == 'World') 
                                         & (renewable_energy_prod_df['Year'] == 2021)]
world_prod_df2 = world_prod_df.drop(columns = ['Entity', 'Year'])
world_prod_df2
````

Drop "Entity" and "Year" column so that pie chart can be plotted with world_prod_df2.iloc[0] which only contain data for each renewable energy type, <br>autopct='%1.1f%%' was put in to generate percentage on the pir chart

````python
energy_type = ['Wind', 'Hydro', 'Solar', 'Others']
plt.figure(figsize = (10,10));
plt.title("Distribution of Renewable Energy of the World in 2021")
plt.pie(world_prod_df2.iloc[0], labels = energy_type, autopct='%1.1f%%',
        colors = ["limegreen", "royalblue", "orangered", "darkviolet"]);
plt.legend(energy_type);
````

<h3>From the pie chart above, It can be conclude that in the year 2021, Hydro energy generate the most electricity out of the other type of renewable energies in the world, with 54% in total. <br>
    Next up is wind energy with 23.3%, follow by solar at 13.1% and other type at 9.6%

### Q3: What is the trend of renewable energy in the world from 1965 to 2021 (use line graph)

First, create a dataframe that only take 'World' from "Entity" column

````python
world_prod_trend_df = renewable_energy_prod_df[renewable_energy_prod_df['Entity'] == 'World']
# world_prod_trend_df.drop(columns= 'Entity', inplace=True)
# world_prod_trend_df.reset_index(drop=True, inplace=True)
# world_prod_trend_df
world_prod_trend_df
````

**NOTE** when trying to create a line graph with the following code, it will generate an **ERROR** <br>
Thus its better to convert them to numpy array <br>
Using plt.plot(world_prod_trend_df['Year'], world_prod_trend_df['Electricity from wind (TWh)']) <br>
will result in an error

Convert the data into numpy array for line graph. 

````python
world_trend_year = world_prod_trend_df["Year"].to_numpy()
world_trend_wind = world_prod_trend_df["Electricity from wind (TWh)"].to_numpy()
world_trend_hydro = world_prod_trend_df["Electricity from hydro (TWh)"].to_numpy()
world_trend_solar = world_prod_trend_df["Electricity from solar (TWh)"].to_numpy()
world_trend_other = world_prod_trend_df["Electricity from other renewables including bioenergy (TWh)"].to_numpy() 
````

x axis will be years and y axis will be type of energy, multiple line graph will be plotted to see the comparison 

````python
plt.plot(world_trend_year, world_trend_wind, label = 'Wind', color = 'limegreen')
plt.plot(world_trend_year, world_trend_hydro, label = 'Hydro', color = 'royalblue')
plt.plot(world_trend_year, world_trend_solar, label = 'Solar', color = 'orangered')
plt.plot(world_trend_year, world_trend_other, label = 'Other', color = 'darkviolet')
plt.legend()
plt.xlabel("Year 1965 to 2021")
plt.ylabel("Energy produced (TWh)")
plt.title("The renewable energy production trend of the world\n from year 1965 to 2021");
````

<h3> From the line graph above, hydro energy had been dominant since 1965, with 923.198TWh produced whereas other type of renewable energy wasn't develop yet<br><br>
    From 1980, other type of renewable energy was already starting to rise, and only at 2000 did solar and wind energy start to develop <br><br>
    After 2010, wind energy starts to overtake other type of renewable energy such as bio and geothermal and around 2018, solar energy starts to overtake it as well. <br><br>
    It can be observe that wind energy are rising fast but compare to hydro energy, it stills produce the most energy in the world

### Q4: Which continent generate the most energy in 2020 (Disregard hydro energy as its too high compare to others)

Create a dataframe that only had continent data and in the year 2020, drop the hydro energy as well as this question wants to see the comparison of other types of renewable energy<br>
A total energy column was created with the total from the 3 type of renewable energy column

````python
continent_df = renewable_energy_prod_df[((renewable_energy_prod_df['Entity'] == 'Asia') |
                                       (renewable_energy_prod_df['Entity'] == 'Eastern Africa') |
                                       (renewable_energy_prod_df['Entity'] == 'Western Africa') |
                                       (renewable_energy_prod_df['Entity'] == 'North America') |
                                       (renewable_energy_prod_df['Entity'] == 'South America') |
                                       (renewable_energy_prod_df['Entity'] == 'Europe') |
                                       (renewable_energy_prod_df['Entity'] == 'Oceania')) &
                                       (renewable_energy_prod_df['Year'] == 2020)]
continent_df2 = continent_df.drop(columns='Electricity from hydro (TWh)')
continent_df2['Total energy (TWh)'] = continent_df2['Electricity from wind (TWh)'] + continent_df2['Electricity from solar (TWh)'] + continent_df2['Electricity from other renewables including bioenergy (TWh)']
continent_df2
````

Use the barplot from seaborn to plot

````python
sns.barplot(x='Entity', y='Total energy (TWh)', data=continent_df2)
plt.title('Energy Distribution Different Continent (exclude hydro energy)');
````

<h3> From the bar chart above, Asia produce the most renewable energy out of the other continent in the world and Western Africa produce the least
